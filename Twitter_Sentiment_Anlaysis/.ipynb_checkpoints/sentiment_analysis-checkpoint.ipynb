### Description: Sentiment analysis program to parse tweets fteched from twitter

##### Importing libraries

```python
import pandas as pd
import numpy as np
import re
import matplotlib.pyplot as plt
import string
import nltk
import seaborn as sns
from nltk.tokenize import word_tokenize
import warnings

%matplotlib inline

warnings.filterwarnings('ignore')
plt.style.use("fivethirtyeight")
```

##### Loading dataset

```python
df = pd.read_csv("training_data.csv")
df.head()
```

```python
df.info()
```

##### Preprocessing dataset

```python
# Removes pattern from a given input

def remove_pattern(input_txt, pattern):
    r = re.findall(pattern, input_txt)
    for word in r:
        input_txt = re.sub(word, "", input_txt)

    return input_txt
```

```python
# Remove twitter username (@user) from text

df["clean_tweet"] = np.vectorize(remove_pattern)(df["tweet"], "@[\w]*")

df.head()
```

```python
# Remove special characters, numbers and punctuations

df['clean_tweet'] = df['clean_tweet'].str.replace("[^a-zA-Z#]", " ")
df.head()
```

```python
# Removing short words

df['clean_tweet'] = df['clean_tweet'].apply(lambda x: " ".join([w for w in x.split() if len(w) > 3]))
df.head()
```

```python
# Individual words considered as tokens

tokenized_tweet = df['clean_tweet'].apply(lambda x: x.split(" "))
tokenized_tweet.head()
```

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()

# Stemming the words
tokenized_tweet = tokenized_tweet.apply(lambda sentence: [stemmer.stem(word) for word in sentence])
tokenized_tweet.head()
```

```python
# Convert the tokenized words into sentences

for i in range(len(tokenized_tweet)):
    tokenized_tweet[i] = " ".join(tokenized_tweet[i])
    
df['clean_tweet'] = tokenized_tweet
df.head()
```

```python
# Extracting the hashtags

def extract_hashtag(tweets):
    hashtags = []
    for tweet in tweets:
        ht = re.findall(r"#(\w+)", tweet)
        hashtags.append(ht)
    
    return hashtags
```

```python
# Extract hashtags from positive tweets
ht_positive = extract_hashtag(df['clean_tweet'][df['label']==0])

# Extract hashtags from negative tweets
ht_negative = extract_hashtag(df['clean_tweet'][df['label']==1])
```

```python
# Unnest / Flatten the lists

ht_positive = sum(ht_positive, [])
ht_negative = sum(ht_negative, [])
```

```python
ht_positive[:5]
```

```python
freq = nltk.FreqDist(ht_positive)
d = pd.DataFrame({'Hashtag': list(freq.keys()),
                  'Count': list(freq.values())})
d.head()
```

```python
# Select the top 10 hashtags

d = d.nlargest(columns="Count", n=10)
plt.figure(figsize=(15,9))
sns.barplot(data=d, x="Hashtag", y="Count")
plt.show()
```

```python
freq = nltk.FreqDist(ht_negative)
d = pd.DataFrame({'Hashtag': list(freq.keys()),
                  'Count': list(freq.values())})
d.head()
```

```python
# Select the top 10 hashtags

d = d.nlargest(columns="Count", n=10)
plt.figure(figsize=(15,9))
sns.barplot(data=d, x="Hashtag", y="Count")
plt.show()
```

##### Input Split

```python
from sklearn.feature_extraction.text import CountVectorizer

bow_vectorizer = CountVectorizer(max_df=0.90, min_df=2, max_features=1000, stop_words="english")
bow = bow_vectorizer.fit_transform(df['clean_tweet'])
```

```python
# bow[0].toarray()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(bow, df["label"], random_state=42, test_size=0.25)
```

##### Model Training

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import f1_score, accuracy_score
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
pred = model.predict(X_test)
f1_score(y_test, pred)
```

```python
accuracy_score(y_test, pred)
```

```python
# Using probablity to get output
pred_prob = model.predict_proba(X_test)
pred = pred_prob[:, 1] >= 0.3
pred = pred.astype(np.int)

f1_score(y_test, pred)
```
